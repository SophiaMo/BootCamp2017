```python
import numpy as np
from matplotlib import pyplot as plt

# Problem 1
def f(n):
    n = np.random.randn(n, n)
    x = np.mean(n, axis = 1)
    y = np.var(x)
    return y
def g():
    a = np.zeros(10)
    for i in range(0, 10):
        a[i] = f((i + 1) * 100)
    plt.plot(a)
    plt.show()
g()
```

```python
# Problem 2
def prob2():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
    plt.plot(x, np.sin(x))
    plt.show()
    plt.plot(x, np.cos(x))
    plt.show()
    plt.plot(x, np.arctan(x))
    plt.show()
prob2()
```

```python
# Problem 3
def prob3():
    x1 = np.linspace(-2, 1-0.00001, 100)
    y1 = [1/ (x - 1) for x in x1]
    x2 = np.linspace(1+0.00001, 6, 150)
    y2 = [1/ (x - 1) for x in x2]
    plt.plot(x1, y1, 'm--', lw = 6)
    plt.plot(x2, y2, 'm--', lw = 6)
    plt.ylim(-6, 6)
    plt.show()
prob3()
```

```python
# Problem 4
def prob4():
    x = np.linspace(0, 2 * np.pi, 100)
    plt.axis([-2, 2, 0, 2 * np.pi,])
    plt.suptitle('sin plots')
    plt.subplot(221)
    plt.plot(x, np.sin(x), 'g-')
    plt.title('sin(x)', fontsize = 10)
    plt.subplot(222)
    plt.plot(x, np.sin(2 * x), 'r--')
    plt.title('sin(2x)', fontsize = 10)
    plt.subplot(223)
    plt.plot(x, 2 * np.sin(x), 'b--')
    plt.title('2sin(x)', fontsize = 10)
    plt.subplot(224)
    plt.plot(x, 2 * np.sin(2 * x), 'm:')
    plt.title('2sin(2x)', fontsize = 10)
    plt.show()
prob4()
```

```python
# Problem 6
def prob6():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
    y = x.copy()
    X, Y = np.meshgrid(x, y)
    Z = np.sin(X) * np.sin(Y) / (X * Y)
    plt.subplot(121)
    plt.pcolormesh(X, Y, Z, cmap="viridis")
    plt.colorbar()
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-2 * np.pi, 2 * np.pi)
    plt.subplot(122)
    plt.contour(X, Y, Z, 10, cmap="Spectral")
    plt.colorbar()
    plt.show()
prob6()
```
