```python
# Problem 2
N = 100
from matplotlib import pyplot as plt
for i in range(6):
    si = np.zeros(N)
    si[1:] = np.random.binomial(1, .5, size=(N-1,))*2-1 
    si = pd.Series(si)
    si = si.cumsum() # random walk
    si.plot()
plt.ylim([-50, 50])
plt.show()

N = 100
si = np.zeros(N)
si[1:] = np.random.binomial(1, .51, size=(N-1,))*2-1 
si = pd.Series(si)
si = si.cumsum() # random walk
si.plot()
plt.ylim([-50, 50])
plt.show()

N = 10000
si = np.zeros(N)
si[1:] = np.random.binomial(1, .51, size=(N-1,))*2-1 
si = pd.Series(si)
si = si.cumsum() # random walk
si.plot()
plt.ylim([-50, 400])
plt.show()

N = 100000
si = np.zeros(N)
si[1:] = np.random.binomial(1, .51, size=(N-1,))*2-1 
si = pd.Series(si)
si = si.cumsum() # random walk
si.plot()
plt.ylim([-50, 3000])
plt.show()
```
