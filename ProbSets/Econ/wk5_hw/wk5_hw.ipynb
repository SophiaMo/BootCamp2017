## Problem 5 and Problem 6 Numerical Solution
Now we would like to find the steady state values for the model specifications in problem 5 and 6 numerically.

```python
def get_cbar(w, r, k, l):
    c_bar = w * l + (r - delta) * k
    return c_bar

def get_rbar(k, l):
    r_bar = alpha * (l/k) ** (1 - alpha) * (np.exp(Zbar)) ** (1 - alpha)
    return r_bar

def get_wbar(k, l):
    w_bar = (1 - alpha) * (k/l) ** alpha * (np.exp(Zbar)) ** (1 - alpha)
    return w_bar
    
def euler_eqns(soln, *args):
    gamma, epsilon, beta, alpha, a, delta, tau, Zbar, problem_5 = args
    if problem_5 is False:
        w = get_wbar(soln[0], soln[1])
        r = get_rbar(soln[0], soln[1])
        c = get_cbar(w, r, soln[0], soln[1])
        eq1 = c ** (-gamma) - beta * (c ** (-gamma)* ((1 - tau) * (r - delta) + 1))
        eq2 = a * (1 - soln[1]) ** (-epsilon) - c ** (-gamma) * w * (1 - tau)  
        errors = np.array([eq1, eq2])
    else:
        w = get_wbar(soln, 1)
        r = get_rbar(soln, 1)
        c = get_cbar(w, r, soln, 1)
        errors = c ** (-gamma) - beta * (c ** (-gamma)* ((1 - tau) * (r - delta) + 1))
    return errors

def get_Tbar(w, r, k, l):
    T_bar = tau * (w * l + (r - delta) * k)
    return T_bar

def get_ibar(k):
    i_bar = delta * k
    return i_bar

def get_ybar(k, l):
    y_bar = k ** alpha * l ** (1 - alpha)
    return y_bar
```

```python
args1 = (gamma, epsilon, beta, alpha, a, delta, tau, Zbar, True)
args2 = (gamma, epsilon, beta, alpha, a, delta, tau, Zbar, False)
result1 = opt.fsolve(euler_eqns, 1, args = (args1))
w1 = get_wbar(result1[0], 1)
r1 = get_rbar(result1[0], 1)
T1 = get_Tbar(w1, r1, result1[0], 1)
i1 = get_ibar(result1[0])
y1 = get_ybar(result1[0], 1)
result2 = opt.fsolve(euler_eqns, np.array([1, 0.8]), args = (args2))
k2, l2 = result2
w2 = get_wbar(result2[0], result2[1])
r2 = get_rbar(result2[0], result2[1])
c2 = get_cbar(w2, r2, k2, l2)
T2 = get_Tbar(w2, r2, result2[0], result2[1])
i2 = get_ibar(result2[0])
y2 = get_ybar(result2[0], result2[1])

print("steady-state capital for problem 5 is: "+str(result1[0]))
print("steady-state wage for problem 5 is: "+str(w1))
print("steady-state interest rate for problem 5 is: "+str(r1))
print("steady-state consumption for problem 5 is: "+str(get_cbar(w1, r1, result1[0], 1)))
print("steady-state government transfer for problem 5 is: "+str(T1))
print("steady-state investment for problem 5 is: "+str(i1))
print("steady-state output for problem 5 is: "+str(y1))
print("")
print("steady-state capital and labor for problem 6 are: "+str(result2))
print("steady-state wage for problem 6 is: "+str(w2))
print("steady-state interest rate for problem 6 is: "+str(r2))
print("steady-state consumption for problem 6 is: "+str(c2))
print("steady-state government transfer for problem 6 is: "+str(T2))
print("steady-state investment for problem 6 is: "+str(i2))
print("steady-state output for problem 6 is: "+str(y2))
```

```python
# Value Function Iteration
VFtol = 1e-6
VFdist = 7.0
VFmaxiter = 3000
V = np.zeros((sizek, sizez))  # initial guess of value function
Vmat = np.zeros((sizek, sizek, sizez))  # initialize Vmat matrix
VFiter = 1

start = time.time()
while VFdist > VFtol and VFiter < VFmaxiter:
    TV = V
    for i in range(sizek):  # loop over k
        for j in range(sizek):  # loop over k'
            for k in range(sizez): # loop over z
                Vmat[i, j, k] = u[i, j, k] + beta * np.dot(V[j, :], pi[:, k])
    
    # iteration for graphing later
    V = Vmat.max(axis=1)  # apply max operator to Vmat (to get V(k))
    PF = np.argmax(Vmat, axis=1)  # find the index of the optimal k'
    VFdist = (np.absolute(V - TV)).max()  # check distance between value
    # function for this iteration and value function from past iteration
    VFiter += 1
end = time.time()

if VFiter < VFmaxiter:
    print('Value function converged after this many iterations:', VFiter)
    print('Time taken for convergence:' + str(end - start))
else:
    print('Value function did not converge')
```

To linearize, we would like to write a gamma function in terms of the state variables, based on the Euler Equation, first. The Euler equation for the Brock-Mirman Model is:  
$\frac{1}{e^{z_t}K_t^\alpha - K_{t+1}} = \beta E_t\{\frac{\alpha e^{z_{t+1}}K_{t+1}^{\alpha-1}}{e^{z_{t+1}}K_{t+1}^{\alpha}-K_{t+2}}\}$， so by rewritting the associated Euler equation, we get:  
$\Gamma(K_{t+2}, K_{t+1}, K_{t}, z_{t+1}, z_{t}) = \frac{1}{e^{z_t}K_t^\alpha - K_{t+1}} = \beta E_t\{\frac{\alpha e^{z_{t+1}}K_{t+1}^{\alpha-1}}{e^{z_{t+1}}K_{t+1}^{\alpha}-K_{t+2}}\}$

```python
def Modeldyn(theta0, params):
    '''
    This function takes vectors of endogenous and exogenous state variables
    along with a vector of 'jump' variables and returns values from the
    characterizing Euler equations.
    
    Inputs are:
        theta: a vector containng (Kpp, Kp, K, Zp, Z) where:
            Kpp: value of capital in two periods
            Kp: value of capital in next period
            K: value of capital this period
            Zp: value of productivity in next period
            Z: value of productivity this period
        params: list of parameter values
    
    Output are:
        Euler: an associated Euler equation rewritten so that they are zero at the
            steady state values of X, Y & Z.  
    '''
    
    # unpack theat0
    (Kpp, Kp, K, Zp, Z) = theta0
    
    E = 1/(np.exp(Z) * K ** alpha - Kp) - beta * ((alpha * np.exp(Zp) * Kp ** (alpha - 1))/ 
                                             (np.exp(Zp) * Kp ** alpha - Kpp))
                                                                            
    return E
```

		[(FP+Q)P+H]\tilde{X}_{t-1}+[(FQ+L)N+(FP+G)Q+M]\tilde{Z}_t &=0
\end{align*}

## Problem 4
Here we solve for the steady-state values using the LinApp functions, by letting the $k_{t-1}$ be the state variables.

```python
def get_c(w, r, kp, k, l):
    c = w * l + (r - delta) * k + k - kp
    return c

def get_r(k, l, z):
    r = alpha * (l/k) ** (1 - alpha) * (np.exp(z)) ** (1 - alpha)
    return r

def get_w(k, l, z):
    w = (1 - alpha) * (k/l) ** alpha * (np.exp(z)) ** (1 - alpha)
    return w
```

```python
def Modeldyn(theta0, params):
    # unpack theat0
    (Xpp, Xp, X, Yp, Y, Zp, Z) = theta0
    w = get_w(X, Y, Z)
    wp = get_w(Xp, Yp, Zp)
    r = get_r(X, Y, Z)
    rp = get_r(Xp, Yp, Zp)
    c = get_c(w, r, Xp, X, Y)
    cp = get_c(wp, rp, Xpp, Xp, Yp)
    # Evaluate Euler equations
    E1 = a * (1 - Y) ** (-epsilon) - (c ** (-gamma) * w * (1 - tau)) 
    E2 = c ** (-gamma)  - beta * (cp ** (-gamma)* ((1 - tau) * (rp - delta) + 1))
    
    return np.array([E1, E2])
```

In the notation of DSGE models, the state variables for an OLG model with 3 generations are  

$$X_t = (k_{2, t+1}, k_{3, t+1})$$

```python
# Household Parameters
nvec = np.array([1.0,1.0,0.2]) # labor supply
yrs_live = 60
s = 3
beta = 0.442
sigma = 3

# Firm Parameters
alpha = 0.35
delta_annual = 0.05
delta = 1- ((1 - delta_annual) ** (yrs_live/s))
```

```python
# Define functions
def get_L(nvec): # function for aggregate labor
    L = nvec.sum()
    return L

def get_K(kvec): # function for aggregate capital
    K = kvec.sum()
    return K

def get_w(K, L, z): # function for wage
    w = (1 - alpha) * np.exp(z) * ((K / L) ** alpha)
    return w

def get_r(K, L, z): # function for interest rate
    r = alpha * np.exp(z) * ((L / K) ** (1 - alpha)) - delta
    return r
```

```python
def Model_olg(theta0, params):
    # unpack theat0
    (k2pp, k3pp, k2p, k3p, k2, k3, zp, z) = theta0
    
    Kp = get_K(np.array([k2p, k3p]))
    K = get_K(np.array([k2, k3]))
    L = get_L(nvec)
    
    wp = get_w(Kp, L, zp)
    w = get_w(K, L, z)
    rp = get_r(Kp, L, zp)
    r = get_r(K, L, z)
    
    c1 = w * nvec[0] - k2
    c2 = (1 + r) * k2 + w * nvec[1] - k3
    c2p = (1 + rp) * k2p + wp * nvec[1] - k3p
    c3p = (1 + rp) * k3p + wp * nvec[2]
    
    # Evaluate Euler equations
    E1 = c1 ** (-sigma) - beta * (1 + rp) * c2p ** (-sigma)
    E2 = c2 ** (-sigma) - beta * (1 + rp) * c3p ** (-sigma)
    
    return np.array([E1, E2])
    
```

```python
# make parameter list to pass to functions
params = (nvec, beta, sigma, alpha)

# set LinApp parameters
nx = 2
ny = 0
nz = 1
zbar = np.array([0.])
```

```python
# take a guess for steady state values of k and ell
guessk = np.array([.02, .04])

# find the steady state values using LinApp_FindSS
k2bar, k3bar = LinApp_FindSS(Model_olg, params, guessk, zbar, nx, ny)
print ('k2bar: ', k2bar)
print ('k3bar: ', k3bar)

# set up steady state input vector
theta0 = np.array([k2bar, k3bar, k2bar, k3bar, k2bar, k3bar, 0., 0.])

# check SS solution
check = Model_olg(theta0, params)
print ('check: ', check)
if np.max(np.abs(check)) > 1.E-6:
    print ('Have NOT found steady state')
```

This is exactly the same as we got for the steady-state values of savings, using TPI.

```python
# find the derivatives matrices
[AA, BB, CC, DD, FF, GG, HH, JJ, KK, LL, MM, WW, TT] = \
    LinApp_Deriv(Model_olg, params, theta0, nx, ny, nz, logX = True)
print('FF: ', FF)
print('GG: ', GG)
print('HH: ', HH)
print('LL: ', LL)
print('MM: ', MM)

# set value for NN    
NN = 1
    
# find the policy and jump function coefficients
PP, QQ, UU, RR, SS, VV = \
    LinApp_Solve(AA,BB,CC,DD,FF,GG,HH,JJ,KK,LL,MM,WW,TT,NN,zbar,Sylv=1)
print ('PP:',PP)
print ('QQ', QQ)
```

```python
k2_start = np.log(0.8 * k2bar) - np.log(k2bar)
k3_start = np.log(1.1 * k3bar) - np.log(k3bar)

kmat = np.zeros((2, 20)) # matrix for k2, k3

kmat[0, 0] = k2_start
kmat[1, 0] = k3_start

for i in range(1, 20):
    kmat[:, i] = np.dot(PP, kmat[:, i-1]) # we don't need Q since Z is fixed
kmat[0, :] = k2bar * np.exp(kmat[0, :]) 
kmat[1, :] = k3bar * np.exp(kmat[1, :]) 

k_agg = kmat[0, :] + kmat[1, :]
```

```python
%matplotlib inline
pylab.rcParams['figure.figsize'] = (10, 5)
plt.subplot(221)
plt.plot(np.arange(0, 20), kmat[0, :])
plt.title('Time Path of K2')
plt.subplot(222)
plt.plot(np.arange(0, 20), kmat[1, :])
plt.title('Time Path of K3')
plt.subplot(223)
plt.plot(np.arange(0, 20), k_agg)
plt.title('Time Path of Aggregate Capital')
plt.tight_layout()
plt.show()
```

The plots we obtain are similar to the ones generated using TPI. One slight difference is that the aggregate capital seems to oscillate more before settling back to steady state values. Log-linearization is much faster than iteration, however we loss some accuracy since we are trying to linearize around the steady-state as opposed to calculate the precise values. However, this loss in accuracy does not hinder our ability to understand the general trends going on in the economy (how capital levels reach steady state). Notably, the time taken for aggregate capital to be close to steady-state level, as predicted by log-linearization, is also very similar to the one predicted by TPI (about 7 periods).
